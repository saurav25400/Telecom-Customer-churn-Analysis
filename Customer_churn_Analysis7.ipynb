```python
from pandas._libs.tslibs.offsets import CustomBusinessHour
# Q2.Which contract type experiences the highest customer churn?
# df["Contract"].head()
churned_customers=df[df["Churn"]=="Yes"]
churned_customers=churned_customers.groupby("Contract").size().reset_index(name="Churned_Customers_count")
churned_customers
print("*"*50)

# plotting column chart for representing different contract Types count

plt.bar(
    churned_customers["Contract"],
    churned_customers["Churned_Customers_count"]
)
plt.xlabel("Contract")
plt.ylabel("Churned_Customers_count")
plt.title("Churned_Customers_count")
plt.show()
```

```python
# Q3.Does customer tenure influence churn behavior?
# df.head()
churned_customer=df[df['Churn']=="Yes"]
customer_tenure=churned_customer.groupby(["Tenure Band"]).size().reset_index(name="churn_Customer_count")
customer_tenure
# churned_customer
# plotting the bar graph for New Customer vs Exisitng customer
plt.bar(
    customer_tenure["Tenure Band"],
    customer_tenure["churn_Customer_count"],
    color="orange"
)
plt.xlabel("Tenure Band")
plt.ylabel("churn_Customer_count")
plt.title("churn Customer count")
plt.show()
```

```python
# Q4.How do monthly charges and total charges impact churn?
# df.head()
monthly_total_charges=df.groupby("Churn")[["MonthlyCharges","TotalCharges"]].mean().reset_index()
monthly_total_charges


```

```python
#  box plot for both Average Monthly Charges and Averaeg TotalCharges
fig, ax = plt.subplots(1, 2, figsize=(14, 6))
sns.boxplot(x="Churn", y="MonthlyCharges", data=df, ax=ax[0])
ax[0].set_title("Monthly Charges by Churn")
ax[0].set_xlabel("Churn")
ax[0].set_ylabel("Monthly Charges")
sns.boxplot(x="Churn", y="TotalCharges", data=df, ax=ax[1])
ax[1].set_title("Total Charges by Churn")
ax[1].set_xlabel("Churn")
ax[1].set_ylabel("Total Charges")
plt.tight_layout()

```

```python
# Q5. Which payment methods are associated with the highest churn?
# df.head()
payment_methods_count=df.groupby(["PaymentMethod", "Churn"])["Churn"].size().unstack(fill_value=0)
payment_methods_count["Churn_Rate"]=payment_methods_count["Yes"]/(payment_methods_count["Yes"]+payment_methods_count["No"])*100
payment_methods_count=payment_methods_count.round(2)
payment_methods_count
```

```python
# Q6 Does internet service type influence customer churn?
# df.head()
internet_service_influence=df.groupby(["InternetService", "Churn"])["Churn"].size().unstack(fill_value=0)
internet_service_influence["Churn_Rate"]=internet_service_influence["Yes"]/(internet_service_influence["Yes"]+internet_service_influence["No"])*100
internet_service_influence=internet_service_influence.round(2)
internet_service_influence = internet_service_influence.reset_index()

internet_service_influence

print("*"*1000)

# internet_service_influence["Churn_Rate"].plot(
#     kind="bar",
#     figsize=(10,6),
#     color="steelblue"
# )
sns.barplot(data=internet_service_influence,x="InternetService",y="Churn_Rate",color="steelblue")

plt.xlabel("Internet Service")
plt.ylabel("Churn Rate (%)")
plt.title("Customer Churn Rate by Internet Service")
plt.xticks(rotation=45)
plt.show()
```

```python
# Q7.Are senior citizens more likely to churn than other customers?
# df.head()
churned_analysis_with_citizens=df.groupby(["SeniorCitizen", "Churn"])["Churn"].size().unstack(fill_value=0)
churned_analysis_with_citizens["Churn Rate (%)"]= churned_analysis_with_citizens["Yes"]/(churned_analysis_with_citizens["Yes"]+churned_analysis_with_citizens["No"])*100
churned_analysis_with_citizens=churned_analysis_with_citizens.round(2)
churned_analysis_with_citizens = churned_analysis_with_citizens.reset_index()
churned_analysis_with_citizens
sns.barplot(data=churned_analysis_with_citizens,x="SeniorCitizen",y="Churn Rate (%)",color="steelblue")
plt.xlabel("Senior Citizen")
plt.ylabel("Churn Rate (%)")
plt.title("Customer Churn Rate by Senior Citizen")
# plt.xticks(rotation=45)
plt.xticks([0, 1], ["No", "Yes"])
plt.show()
```

```python
#Q9. Which customer risk category contributes the highest number of churned customers?
# df.head()
# churend_customer=df[df["Churn"]=="Yes"]
risk_analysis_customers=df[df["Churn"]=="Yes"].groupby("Risk Type").size().reset_index(name="churend_customers")
risk_analysis_customers=risk_analysis_customers.sort_values(by="churend_customers",ascending=False)
risk_analysis_customers

# barplot plotting for churend customers
sns.barplot(data=risk_analysis_customers,x="Risk Type",y="churend_customers",color="steelblue", palette="viridis")
plt.xlabel("Risk Type")
plt.ylabel("Churend_customers")
plt.title("Risk Type by Churend_customers ")
```
